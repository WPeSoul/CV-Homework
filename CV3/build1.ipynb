```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取图像
img1 = cv2.imread('building1.jpg')
img2 = cv2.imread('building2.jpg')
img3 = cv2.imread('building3.jpg')
img4 = cv2.imread('building4.jpg')
img5 = cv2.imread('building5.jpg')

# 创建一个图像列表
images = [img1, img2, img3, img4, img5]

# 创建Stitcher对象
stitcher = cv2.Stitcher_create()

# 尝试拼接图像
status, panorama = stitcher.stitch(images)

if status == cv2.Stitcher_OK:
    # 拼接成功，使用matplotlib显示结果
    # 注意：OpenCV图像是BGR格式的，而matplotlib期望的是RGB格式
    panorama_rgb = cv2.cvtColor(panorama, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize=(panorama.shape[1] / 100, panorama.shape[0] / 100))  # 调整图像大小以适应显示
    plt.imshow(panorama_rgb)
    plt.title('Panorama')
    plt.axis('off')  # 关闭坐标轴
    plt.show()
    
    # 如果你还想保存结果，可以使用cv2.imwrite
    # cv2.imwrite('panorama.jpg', panorama)
    print("全景图拼接成功")
else:
    print("全景图拼接失败，错误代码:", status)
```
