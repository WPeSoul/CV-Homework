```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 
img1 = cv2.imread('dsc_2705.jpg')
img2 = cv2.imread('dsc_2706.jpg')
img3 = cv2.imread('dsc_2707.jpg')
img4 = cv2.imread('dsc_2708.jpg')

# 
images = [img2, img3, img4, img1]

# Stitcher
stitcher = cv2.Stitcher_create()

# 
status, panorama = stitcher.stitch(images)

if status == cv2.Stitcher_OK:
    # matplotlib
    # OpenCVBGRmatplotlibRGB
    panorama_rgb = cv2.cvtColor(panorama, cv2.COLOR_BGR2RGB)
    
    plt.figure(figsize=(panorama.shape[1] / 100, panorama.shape[0] / 100))  # 
    plt.imshow(panorama_rgb)
    plt.title('Panorama')
    plt.axis('off')  # 
    plt.show()
    
    # cv2.imwrite
    # cv2.imwrite('panorama.jpg', panorama)
    print("")
else:
    print(":", status)
```
